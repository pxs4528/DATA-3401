```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

x = np.linspace(-1*2*np.pi,2*np.pi,num=100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = []
for i in x:
    if i >= 0:
        y3.append(i ** 2)
    elif i < 0:
        y3.append(0)
plt.plot(x,y1,x,y2,x,y3)
print(y3)
plt.show()
```
