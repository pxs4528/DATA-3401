## Exercise 1

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

x = np.linspace(-1*2*np.pi,2*np.pi,num=100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = []
for i in x:
    if i >= 0:
        y3.append(i ** 2)
    elif i < 0:
        y3.append(0)
plt.plot(x,y1, color='cyan', linestyle='dotted', label= 'sinx')
plt.plot(x,y2, color='green', linestyle='dashed', label= 'cosx')
plt.plot(x,y3, color='magenta', linestyle='solid', label= 'function')
leg = plt.legend(loc='upper center')
plt.xlabel("-2pi to 2pi")
plt.ylabel("y")
plt.title("Exercise 1")
plt.show()
```

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

x = np.linspace(-1,1,num=100)
y = np.linspace(-1,1,num=100)
X,Y= np.meshgrid(x,y)
lst = np.sin(10*(X ** 2 + Y ** 2))/10
print(type(lst))
fig, axes = plt.subplots()
color = matplotlib.cm.summer
contour = axes.contour(lst, cmap= color, levels= 20)
fig.colorbar(contour,ax=axes)
```
