```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

x = np.linspace(-1*2*np.pi,2*np.pi,num=100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = []
for i in x:
    if i >= 0:
        y3.append(i ** 2)
    elif i < 0:
        y3.append(0)
plt.plot(x,y1,x,y2,x,y3)
print(y3)
plt.show()
```

```python
x = np.linspace(-1,1,num=100)
y = np.linspace(-1,1,num=100)
X,Y= np.meshgrid(x,y)
lst = np.sin(10*(X ** 2 + Y ** 2))/10
print(type(lst))
fig, axes = plt.subplots()
color = matplotlib.cm.cool
contour = axes.contour(lst, cmap= color, levels= 20)
fig.colorbar(contour,ax=axes)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

x = np.linspace(-1,1,num=100)
y = np.linspace(-1,1,num=100)
X,Y= np.meshgrid(x,y)
lst = np.sin(10*(X ** 2 + Y ** 2))/10
print(type(lst))
fig = plt.figure(figsize = (14,6))
                 
ax = fig.add_subplot(1,2,1, projection = '3d')

p = ax.plot_surface(X,Y,lst,rstride=1,cstride=1,antialiased=False)#,rstride=4,cstride=4,linewidth=0)

ax = fig.add_subplot(1,2,2,projection='3d')
p = ax.plot_surface(X,Y,lst,cmap = color,rstride=1,cstride=1,antialiased=True)#,rstride=1,cstride=1,cmap = color,antialiased=True)
ax.view_init(1000,20)
cb = fig.colorbar(p,shrink=0.5)
```
